```python
def switch_last_updated(df, col, newcol):
    df[newcol] = df[col]
    for i in range(0,len(df[col])):
        if df[col][i] == "today":
            df[newcol][i] = 1        
        elif df[col][i] == "yesterday":
            df[newcol][i] = 2
        elif df[col][i] == "a week ago" or df[col][i] == "1 week ago" :
            df[newcol][i] = 7
        else:
            for j in range(1,58):
                if df[col][i] == str(j) + " months ago":
                    df[newcol][i] = j * 30                
                elif df[col][i] == str(j) + " weeks ago":
                    df[newcol][i] = j * 7
                elif df[col][i] == str(j) + " days ago":
                    df[newcol][i] = j
       
        
    return df
df = switch_last_updated(df,"property_last_updated", "property_last_updated_num")
```
